```python
#this cell is used to proeed the .DAT file of each airfoil. 
#fisrt of all, the length of coordinate ot each airfoil is different, for these different airfoil length, we would like to make sure all the same length to 200. How to make it? interploation method will be used to reach this target. 

# 导入必要的库
import os
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

# 文件路径
folder_path = "/workspace/ljl/Airfoil_aerodynamics/Coordinate"  # 替换为包含.DAT文件的文件夹路径
folder_path_new = '/workspace/ljl/Airfoil_aerodynamics/Interploated_coordinates'  #新的路径保存插值之后的坐标文件

# 遍历文件夹中的文件
for filename in os.listdir(folder_path):
    if filename.endswith(".DAT"):
        file_path = os.path.join(folder_path, filename)
        
        # 打开文件并读取内容
        with open(file_path, "r") as file:
            # 读取第一行（机翼型号）
            wing_model = file.readline().strip()
            
            # 读取剩余行（坐标）
            coordinates = file.readlines()
            
            # 处理机翼坐标数据
            # 这里可以根据需要对坐标数据进行处理或存储
            
            # 提取上表面和下表面的坐标
            half_len = len(coordinates) // 2
            upper_surface = coordinates[:half_len]
            lower_surface = coordinates[half_len:]
            
            # 提取上表面的x坐标和y坐标
            upper_x = [float(coord.split()[0]) for coord in upper_surface]
            upper_y = [float(coord.split()[1]) for coord in upper_surface]
            
            # 提取下表面的x坐标和y坐标
            lower_x = [float(coord.split()[0]) for coord in lower_surface]
            lower_y = [float(coord.split()[1]) for coord in lower_surface]
            
            # 创建上表面和下表面的插值函数
            upper_interp_func = interp1d(upper_x, upper_y, kind='linear')
            lower_interp_func = interp1d(lower_x, lower_y, kind='linear')
            
            # 生成新的坐标点（上表面）
            new_x_upper = np.linspace(min(upper_x), max(upper_x), 200)
            new_upper_y = upper_interp_func(new_x_upper)
            
            # 生成新的坐标点（下表面）
            new_x_lower = np.linspace(min(lower_x), max(lower_x), 200)
            new_lower_y = lower_interp_func(new_x_lower)

            # 保存上表面和下表面的坐标
            output_path = os.path.join(folder_path_new, f"interpolated_{filename}")
            with open(output_path, "w") as output_file:
                for i in range(len(new_x_upper)-1, -1, -1):
                    output_file.write(f"{new_x_upper[i]} {new_upper_y[i]}\n")
                
                for i in range(len(new_x_lower)):
                    output_file.write(f"{new_x_lower[i]} {new_lower_y[i]}\n")
            '''
            # 绘制机翼图像
            plt.figure(figsize=(8,6), dpi=200)
            
            # 绘制上半部分
            plt.plot(new_x_upper, new_upper_y, 'b-')
            
            # 绘制下半部分
            plt.plot(new_x_lower, new_lower_y, 'b-')
            
            #plt.legend()
            plt.xlim([0,1])
            plt.ylim([-0.5, 0.5])
            plt.xlabel('X')
            plt.ylabel('Y')
            plt.title('Wing Plot: ' + wing_model)
            plt.show()
            '''
```
